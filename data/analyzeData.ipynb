```python
import json
from collections import Counter

DATA_PATH = "./gita_clean.json"

with open(DATA_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)

print("ðŸ“Š DATASET OVERVIEW")
print("-" * 40)

# 1. Total verses
total_verses = len(data)
print(f"Total verses: {total_verses}")

# 2. Chapters
chapters = [v["chapter"] for v in data if "chapter" in v]
unique_chapters = sorted(set(chapters))
print(f"Total chapters: {len(unique_chapters)}")
print(f"Chapters present: {unique_chapters}")

# 3. Verses per chapter
chapter_counts = Counter(chapters)
print("\nðŸ“˜ Verses per chapter:")
for ch, count in chapter_counts.items():
    print(f"  Chapter {ch}: {count} verses")

# 4. Missing fields check
required_fields = ["id", "chapter", "verse", "sanskrit", "english"]
missing = {field: 0 for field in required_fields}

for v in data:
    for field in required_fields:
        if field not in v or not v[field]:
            missing[field] += 1

print("\nâŒ Missing field counts:")
for field, count in missing.items():
    print(f"  {field}: {count}")

# 5. English text quality
english_lengths = [len(v["english"].split()) for v in data if "english" in v]

print("\nðŸ“ English text stats:")
print(f"  Avg words per verse: {sum(english_lengths)/len(english_lengths):.2f}")
print(f"  Min words: {min(english_lengths)}")
print(f"  Max words: {max(english_lengths)}")

# 6. Check BG prefixes
bg_prefix_count = sum(
    1 for v in data if v["english"].strip().startswith("BG")
)
print(f"\nðŸ”Ž Verses with 'BG x.y:' prefix: {bg_prefix_count}")

# 7. Duplicate IDs
ids = [v["id"] for v in data]
duplicate_ids = [k for k, v in Counter(ids).items() if v > 1]
print(f"\nðŸ§¯ Duplicate IDs: {len(duplicate_ids)}")
if duplicate_ids:
    print("Duplicates:", duplicate_ids[:5])

print("\nâœ… Analysis complete.")
```
