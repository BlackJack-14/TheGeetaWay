```python
import json
import re
import os
```

```python

RAW_PATH = "./gita_clean.json"
CLEAN_PATH = "./bhagavad_gita_clean_final.json"
```

```python
os.makedirs("../data/processed", exist_ok=True)
with open(RAW_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)

cleaned = []

range_pattern = re.compile(r"^BG\s*(\d+)\.(\d+)-(\d+):\s*")
single_pattern = re.compile(r"^BG\s*\d+\.\d+:\s*")

for v in data:
    english = v["english"].strip()

    # Case 1: multi-verse range (e.g., BG 1.16-18:)
    match = range_pattern.match(english)
    if match:
        chapter, start, end = match.groups()

        # keep ONLY the first verse of the range
        if v["verse"] != int(start):
            continue

        english = range_pattern.sub("", english).strip()

    # Case 2: single verse (e.g., BG 1.1:)
    else:
        english = single_pattern.sub("", english).strip()

    cleaned.append({
        "id": v["id"],
        "chapter": v["chapter"],
        "verse": v["verse"],
        "sanskrit": v["sanskrit"].strip(),
        "english": english
    })

with open(CLEAN_PATH, "w", encoding="utf-8") as f:
    json.dump(cleaned, f, ensure_ascii=False, indent=2)

print(f"✅ Cleaned verses count: {len(cleaned)}")
```
